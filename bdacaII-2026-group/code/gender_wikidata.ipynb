```python
import requests
import json
import pandas as pd
from datetime import datetime
import time

# Define the API endpoint and polite defaults
API_URL = "https://en.wikipedia.org/w/api.php"
USER_AGENT = "MediaWikiAPITutorial/1.0 (contact: your.email@example.com)"
REQUEST_DELAY = 0.5  # seconds between requests
TIMEOUT = 30

session = requests.Session()
session.headers.update({"User-Agent": USER_AGENT})

def api_get(params: dict) -> dict | None:
    """Make a safe request to the MediaWiki API with retries and rate limiting."""
    try:
        response = session.get(API_URL, params=params, timeout=TIMEOUT)
        response.raise_for_status()
        data = response.json()
    except requests.exceptions.RequestException as e:
        print(f"Request error: {e}")
        return None
    except json.JSONDecodeError:
        print("Response was not valid JSON")
        return None
    finally:
        time.sleep(REQUEST_DELAY)

    return data
```

```python
def get_wikidata_id(title):
    """
    Retrieve the full text content of a Wikipedia article.
    
    Parameters:
    title (str): The title of the Wikipedia article
    
    Returns:
    str: The article content in plain text
    """
    params = {
        'action': 'query',
        'format': 'json',
        'titles': title,
        'prop': 'pageprops',  #
        'ppprop':'wikibase_item',  # Get the Wikidata ID
        "format":"json"
    }
    
    data = api_get(params)
    if not data or 'query' not in data:
        return "Request failed or no data"
    print(data)
    return data

# Example: Get content about "Climate change"
article_title = "Climate change"
content = get_wikidata_id(article_title)
```
